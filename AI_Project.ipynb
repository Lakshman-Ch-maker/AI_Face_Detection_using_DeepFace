- ## Now we join every code into a single function named "verify_face()"
- The function takes 2 arguments those are thw two image paths

```python
def verify_face(img1_path, img2_path):
    
    img1= cv2.imread(img1_path)
    img2= cv2.imread(img2_path)
    
    plt.imshow( img1[ : ,  : , : : -1])
    plt.show()
    plt.imshow( img2[ : ,  : , : : -1])
    plt.show()
    
    result = DeepFace.verify(img1_path, img2_path)
    print(" Result: ", result)
    print(result['verified'])
    
    verification = result['verified']
    
    if verification:
        print("both photos have the same person")
    else:
        print("not same person")
```

```python
models = [
  "VGG-Face", 
  "Facenet", 
  "Facenet512", 
  "OpenFace", 
  "DeepFace", 
  "DeepID", 
  "ArcFace", 
  "Dlib", 
  "SFace",
]

#face verification
result = DeepFace.verify(img1_path = "img1.jpg", 
      img2_path = "img2.jpg", 
      model_name = models[1]
)

#face recognition
df = DeepFace.find(img_path = "img1.jpg",
      db_path = "D:\exp\AI_project", 
      model_name = models[1]
)

#embeddings
embedding = DeepFace.represent(img_path = "img1.jpg", 
      model_name = models[1]
)
```
